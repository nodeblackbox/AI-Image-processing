## Formulas and Equations

The core of this project lies in its use of CNNs and image processing techniques. Here are some key formulas:

### Convolution Operation
$$ \text{Output} = \sum_{i, j} (\text{Image}_{i,j} \times \text{Filter}_{i,j}) $$
This operation involves sliding a filter over the image and computing the dot product at each position. ðŸ”„

### Pooling Operation
$$ \text{MaxPooling} = \max(\text{Window Elements}) $$
Pooling reduces the spatial dimensions of the image, retaining only the most significant features. ðŸ“‰

### Mean Squared Error (MSE)
$$ \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (Y_{\text{true},i} - Y_{\text{pred},i})^2 $$
MSE is used as the loss function, measuring the average squared difference between actual and predicted heatmaps. ðŸ“Š

### Hand Segmentation Using Color Thresholding ðŸŽ¨âœ‚ï¸

Segmenting the hand from the rest of the image is a pivotal step. Here, we use color thresholding techniques to isolate the hand, preparing it for further analysis.

```python
def ready_the_magnifying_glass(image_path, resize_width=None):
    image = cv2.imread(image_path)
    if resize_width:
        r = resize_width / image.shape[1]
        dim = (resize_width, int(image.shape[0] * r))
        image = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
    image_hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    return image_hsv


def brew_the_potion(image_hsv):
    lower_skin = np.array([0, 48, 80], dtype=np.uint8)
    upper_skin = np.array([20, 255, 255], dtype=np.uint8)
    mask = cv2.inRange(image_hsv, lower_skin, upper_skin)
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)
    mask = cv2.GaussianBlur(mask, (5, 5), 0)
    return mask
```

### Exploring with Sliding Windows ðŸªŸðŸ•µï¸â€â™‚ï¸

To analyze different segments of the image for hand presence, we employ a sliding window technique. This method systematically scans the image, paving the way for effective hand detection.

Next, we implement a sliding window technique to systematically analyze different segments of the image. This step is crucial for identifying regions with higher hand coverage.


### Probability Distribution of Window Coverage ðŸ“ŠðŸ”

After analyzing the image with sliding windows, we plot the probability distribution of hand coverage. This plot helps us understand the frequency and distribution of hand presence across different window sizes.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to slide a window over the image
def sliding_window(image, step_size, window_size):
    for y in range(0, image.shape[0] - window_size[1] + 1, step_size):
        for x in range(0, image.shape[1] - window_size[0] + 1, step_size):
            yield (x, y, image[y:y + window_size[1], x:x + window_size[0]])

# Function to calculate window coverage
def calculate_window_coverage(image, mask, window_sizes, step_size):
    coverage_list = []
    for window_size in window_sizes:
        for (x, y, window) in sliding_window(mask, step_size=step_size, window_size=window_size):
            if window.shape[0] != window_size[1] or window.shape[1] != window_size[0]:
                continue
            hand_coverage = np.sum(window) / (window_size[0] * window_size[1] * 255)
            coverage_list.append(hand_coverage)
    return coverage_list

# Prepare the image and segment the hand
image_hsv = ready_the_magnifying_glass('/content/Hands/Hand_0000004.jpg', resize_width=500)
segmented_hand_mask = brew_the_potion(image_hsv)

# Define window sizes and step size for the sliding window
window_sizes = [(96, 96), (128, 128), (160, 160)]
step_size = 32

# Calculate coverage list
coverage_list = calculate_window_coverage(image_hsv, segmented_hand_mask, window_sizes, step_size)

# Plot the probability distribution of window coverage
plt.figure()
plt.hist(coverage_list, bins=20, color='blue', alpha=0.7)
plt.title('Probability Distribution of Window Coverage')
plt.xlabel('Proportion of Hand within Window')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python

def calculate_heatmaps(image, mask, window_sizes, step_size, threshold=0.5):
    # Initialize heatmaps to zeros
    full_heatmap = np.zeros_like(mask, dtype=np.float32)
    filtered_heatmap = np.zeros_like(mask, dtype=np.float32)

    for window_size in window_sizes:
        for (x, y, window) in sliding_window(mask, step_size=step_size, window_size=window_size):
            if window.shape[0] != window_size[1] or window.shape[1] != window_size[0]:
                continue  # Skip incomplete windows at the image edges

            # Calculate the proportion of the window covered by the hand
            hand_coverage = np.sum(window) / (window_size[0] * window_size[1] * 255)

            # Update full heatmap
            full_heatmap[y:y + window_size[1], x:x + window_size[0]] += hand_coverage

            # Update filtered heatmap only if above threshold
            if hand_coverage >= threshold:
                filtered_heatmap[y:y + window_size[1], x:x + window_size[0]] += hand_coverage

    # Normalize the heatmaps
    full_heatmap /= np.max(full_heatmap)
    filtered_heatmap /= np.max(filtered_heatmap)

    return full_heatmap, filtered_heatmap


# Preparing the image and segmenting the hand
image_hsv = ready_the_magnifying_glass('/content/Hands/Hand_0000004.jpg', resize_width=500)
segmented_hand_mask = brew_the_potion(image_hsv)

# Define the window sizes and step size for the sliding window
window_sizes = [(96, 96), (128, 128), (160, 160)]
step_size = 32

# Generate heatmaps
full_heatmap, filtered_heatmap = calculate_heatmaps(image_hsv, segmented_hand_mask, window_sizes, step_size)

# Convert and display the full heatmap
full_heatmap_jet = cv2.applyColorMap(np.uint8(full_heatmap * 255), cv2.COLORMAP_JET)
full_heatmap_jet = cv2.addWeighted(full_heatmap_jet, 0.5, image_hsv, 0.5, 0)

# Show the full heatmap
plt.imshow(cv2.cvtColor(full_heatmap_jet, cv2.COLOR_BGR2RGB))
plt.title('Full Heatmap')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display


def calculate_continuous_heatmap(image, mask, window_size, step_size, threshold=0.5):
    # Create an accumulator for the heatmap
    heatmap_accumulator = np.zeros_like(mask, dtype=np.float64)

    # Count how many times each pixel is visited
    visit_count = np.zeros_like(mask, dtype=np.float64)

    for (x, y, window) in sliding_window(mask, step_size=step_size, window_size=window_size):
        # Calculate the proportion of the window covered by the hand
        hand_coverage = np.sum(window) / (window_size[0] * window_size[1] * 255)

        if hand_coverage >= threshold:
            # Add the coverage value to the heatmap accumulator
            heatmap_accumulator[y:y + window_size[1], x:x + window_size[0]] += hand_coverage
            # Increment the visit count
            visit_count[y:y + window_size[1], x:x + window_size[0]] += 1

    # Normalize the heatmap by the number of visits
    # Avoid division by zero
    visit_count[visit_count == 0] = 1
    heatmap = heatmap_accumulator / visit_count

    # Normalize the heatmap to be between 0 and 1
    heatmap = heatmap / np.max(heatmap)

    # Apply a colormap to the heatmap (convert to BGR for OpenCV)
    heatmap_colored = cv2.applyColorMap(np.uint8(heatmap * 255), cv2.COLORMAP_JET)

    # Combine with the original image for better visualization
    heatmap_overlay = cv2.addWeighted(heatmap_colored, 0.5, image, 0.5, 0)

    return heatmap_overlay
# Define a function for interactive heatmap generation with adjustable parameters
def interactive_heatmap(window_width, window_height, step_size, threshold):
    window_size = (window_width, window_height)

    # Generate the heatmap using the adjustable parameters
    continuous_heatmap = calculate_continuous_heatmap(
        image_hsv, segmented_hand_mask, window_size, step_size, threshold
    )

    # Display the heatmap
    plt.imshow(cv2.cvtColor(continuous_heatmap, cv2.COLOR_BGR2RGB))
    plt.title(f'Interactive Continuous Heatmap (Threshold: {threshold})')
    plt.show()

# Create interactive sliders for the window size, step size, and threshold
window_width_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Width:')
window_height_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Height:')
step_size_slider = widgets.IntSlider(value=32, min=1, max=50, step=1, description='Step Size:')
threshold_slider = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.01, description='Threshold:')

# Display the interactive elements
interactive_widgets = widgets.interactive(
    interactive_heatmap,
    window_width=window_width_slider,
    window_height=window_height_slider,
    step_size=step_size_slider,
    threshold=threshold_slider
)

display(interactive_widgets)
```

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.losses import MeanSquaredError, MeanAbsoluteError
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
import random

class HeatmapModelTrainer:
    def __init__(self, images_np, heatmaps_np, optimizer='adam', learning_rate=0.001, epochs=10, batch_size=32):
        self.images_np = images_np
        self.heatmaps_np = heatmaps_np
        self.optimizer = optimizer
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        self.model = self.create_model()

    def create_model(self):
        # Adjusted model architecture to be less resource-intensive
        model = Sequential([
            Conv2D(32, (3, 3), activation='relu', input_shape=(320, 240, 1), padding='same'),
            MaxPooling2D((2, 2)),
            Conv2D(64, (3, 3), activation='relu', padding='same'),
            MaxPooling2D((2, 2)),
            Flatten(),
            Dense(1024, activation='relu'),  # Reduced number of neurons
            Reshape((32, 32)),  # Adjusted to a smaller heatmap size
        ])
        optimizer = self.get_optimizer()
        model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=[tf.keras.metrics.MeanSquaredError()])
        return model
    # def __init__(self, images_np, heatmaps_np, model_layers, optimizer='adam', learning_rate=0.001, epochs=10, batch_size=32):
    #     self.images_np = images_np
    #     self.heatmaps_np = heatmaps_np
    #     self.model_layers = model_layers
    #     self.optimizer = optimizer
    #     self.learning_rate = learning_rate
    #     self.epochs = epochs
    #     self.batch_size = batch_size
    #     self.model = self.create_model()




    def process_and_train_ui(self):
        # UI components
        window_width_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Width:')
        window_height_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Height:')
        step_size_slider = widgets.IntSlider(value=32, min=1, max=50, step=1, description='Step Size:')
        threshold_slider = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.01, description='Threshold:')
        learning_rate_slider = widgets.FloatSlider(value=0.01, min=0.0001, max=0.01, step=0.0001, description='Learning Rate:')
        epochs_slider = widgets.IntSlider(value=10, min=1, max=100, step=1, description='Epochs:')
        batch_size_slider = widgets.IntSlider(value=32, min=1, max=128, step=1, description='Batch Size:')
        process_button = widgets.Button(description='Process Images')
        train_button = widgets.Button(description='Train Model')
        interactive_heatmap_button = widgets.Button(description='Generate Interactive Heatmap')
        output = widgets.Output()
        image_output = widgets.Output()

        def on_process_button_clicked(b):
            with output:
                clear_output()
                print("Processing images into heatmaps...")
                self.learning_rate = learning_rate_slider.value
                self.epochs = epochs_slider.value
                self.batch_size = batch_size_slider.value
                self.generate_heatmaps_for_all_images(
                    (window_width_slider.value, window_height_slider.value),
                    step_size_slider.value,
                    threshold_slider.value
                )
                print("Processing complete.")
            self.display_sample_heatmap(image_output)

        def on_train_button_clicked(b):
            with output:
                clear_output()
                print("Training model...")
                history = self.train()
                self.plot_history(history)

        def on_interactive_heatmap_button_clicked(b):
            self.generate_interactive_heatmap(image_output, window_width_slider.value, window_height_slider.value, step_size_slider.value, threshold_slider.value)

        process_button.on_click(on_process_button_clicked)
        train_button.on_click(on_train_button_clicked)
        interactive_heatmap_button.on_click(on_interactive_heatmap_button_clicked)

        ui = widgets.VBox([
            window_width_slider, window_height_slider, step_size_slider, threshold_slider,
            learning_rate_slider, epochs_slider, batch_size_slider,
            process_button, train_button, interactive_heatmap_button, output, image_output
        ])

        display(ui)

    def generate_interactive_heatmap(self, output_area, window_width, window_height, step_size, threshold):
        with output_area:
            clear_output()
            image_dir = '/content/Hands'
            image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg') or f.endswith('.png')]
            if not image_files:
                print("No image files found in the directory")
                return

            random_image_file = random.choice(image_files)
            random_image_path = os.path.join(image_dir, random_image_file)

            image_hsv = ready_the_magnifying_glass(random_image_path, resize_width=None)
            segmented_hand_mask = brew_the_potion(image_hsv)

            continuous_heatmap = self.calculate_continuous_heatmap(segmented_hand_mask, (window_width, window_height), step_size, threshold)

            plt.imshow(continuous_heatmap, cmap='hot')
            plt.title(f'Interactive Continuous Heatmap (Threshold: {threshold})')
            plt.colorbar()
            plt.show()
    def display_sample_heatmap(self, output_area):
        with output_area:
            clear_output()
            sample_index = np.random.randint(0, len(self.images_np))
            sample_image = self.images_np[sample_index].squeeze()
            sample_heatmap = self.heatmaps_np[sample_index].squeeze()

            fig, ax = plt.subplots(1, 2, figsize=(12, 6))
            ax[0].imshow(sample_image, cmap='gray')
            ax[0].set_title('Sample Image')
            ax[1].imshow(sample_heatmap, cmap='hot')
            ax[1].set_title('Generated Heatmap')
            plt.show()

    def get_optimizer(self):
        if self.optimizer.lower() == 'adam':
            return tf.keras.optimizers.Adam(learning_rate=self.learning_rate)
        elif self.optimizer.lower() == 'rmsprop':
            return tf.keras.optimizers.RMSprop(learning_rate=self.learning_rate)
        # Add more optimizers as needed
        else:
            raise ValueError("Unsupported optimizer")

    def split_data(self):
        # Assuming images_np is your images and heatmaps_np is your heatmaps
        train_images, test_val_images, train_heatmaps, test_val_heatmaps = train_test_split(
            self.images_np, self.heatmaps_np, test_size=0.3, random_state=42)
        val_images, test_images, val_heatmaps, test_heatmaps = train_test_split(
            test_val_images, test_val_heatmaps, test_size=0.5, random_state=42)
        return train_images, val_images, test_images, train_heatmaps, val_heatmaps, test_heatmaps

    def train(self):
        train_images, val_images, test_images, train_heatmaps, val_heatmaps, test_heatmaps = self.split_data()
        history = self.model.fit(
            train_images, train_heatmaps,
            validation_data=(val_images, val_heatmaps),
            epochs=self.epochs,
            batch_size=self.batch_size
        )
        return history

    def evaluate(self, images, heatmaps):
        predictions = self.model.predict(images)
        mse = MeanSquaredError()
        mae = MeanAbsoluteError()
        mse_value = mse(predictions, heatmaps).numpy()
        mae_value = mae(predictions, heatmaps).numpy()
        return mse_value, mae_value

    def generate_heatmaps_for_all_images(self, window_size, step_size, threshold):
        self.heatmaps_np = np.array([self.calculate_continuous_heatmap(
            self.images_np[i, :, :, 0], window_size, step_size, threshold
        ) for i in range(self.images_np.shape[0])])

    def plot_history(self, history):
        plt.figure(figsize=(10, 5))
        plt.plot(history.history['mean_squared_error'], label='Train MSE')
        plt.plot(history.history['val_mean_squared_error'], label='Validation MSE')
        plt.title('Model MSE Over Epochs')
        plt.xlabel('Epochs')
        plt.ylabel('Mean Squared Error')
        plt.legend()
        plt.show()

    def calculate_continuous_heatmap(self, mask, window_size, step_size, threshold=0.5):
        # Create an accumulator for the heatmap
        heatmap_accumulator = np.zeros_like(mask, dtype=np.float64)

        # Count how many times each pixel is visited
        visit_count = np.zeros_like(mask, dtype=np.float64)

        for (x, y, window) in sliding_window(mask, step_size=step_size, window_size=window_size):
            # Calculate the proportion of the window covered by the hand
            hand_coverage = np.sum(window) / (window_size[0] * window_size[1] * 255)

            if hand_coverage >= threshold:
                # Add the coverage value to the heatmap accumulator
                heatmap_accumulator[y:y + window_size[1], x:x + window_size[0]] += hand_coverage
                # Increment the visit count
                visit_count[y:y + window_size[1], x:x + window_size[0]] += 1

        # Normalize the heatmap by the number of visits
        visit_count[visit_count == 0] = 1
        heatmap = heatmap_accumulator / visit_count

        # Normalize the heatmap to be between 0 and 1
        heatmap = heatmap / np.max(heatmap)
        return heatmap

    def interactive_heatmap(self, window_width, window_height, step_size, threshold):
        # Directory path containing images
        image_dir = '/content/Hands'

        # List all jpg and png files in the directory
        image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg') or f.endswith('.png')]
        if not image_files:
            print("No image files found in the directory")
            return

        # Select a random image file
        random_image_file = random.choice(image_files)
        random_image_path = os.path.join(image_dir, random_image_file)

        # Read and process the selected image using the file path
        image_hsv = ready_the_magnifying_glass(random_image_path, resize_width=None)  # Pass file path
        segmented_hand_mask = brew_the_potion(image_hsv)

        # Calculate heatmap
        window_size = (window_width, window_height)
        continuous_heatmap = self.calculate_continuous_heatmap(segmented_hand_mask, window_size, step_size, threshold)

        # Display the heatmap
        plt.imshow(continuous_heatmap, cmap='hot')
        plt.title(f'Interactive Continuous Heatmap (Threshold: {threshold})')
        plt.colorbar()
        plt.show()


    def display_interactive_controls(self):
        window_width_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Width:')
        window_height_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Height:')
        step_size_slider = widgets.IntSlider(value=32, min=1, max=50, step=1, description='Step Size:')
        threshold_slider = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.01, description='Threshold:')

        interactive_widgets = widgets.interactive(
            self.interactive_heatmap,
            window_width=window_width_slider,
            window_height=window_height_slider,
            step_size=step_size_slider,
            threshold=threshold_slider
        )
        display(interactive_widgets)

    def interactive_ui(self):
        window_width_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Width:')
        window_height_slider = widgets.IntSlider(value=96, min=10, max=200, step=1, description='Window Height:')
        step_size_slider = widgets.IntSlider(value=32, min=1, max=50, step=1, description='Step Size:')
        threshold_slider = widgets.FloatSlider(value=0.5, min=0, max=1.0, step=0.01, description='Threshold:')
        train_button = widgets.Button(description='Train Model')

        ui = widgets.VBox([
            window_width_slider,
            window_height_slider,
            step_size_slider,
            threshold_slider,
            train_button
        ])

        def on_train_button_clicked(b):
            with output:
                clear_output()
                print("Generating heatmaps...")
                self.generate_heatmaps_for_all_images(
                    (window_width_slider.value, window_height_slider.value),
                    step_size_slider.value,
                    threshold_slider.value
                )
                print("Training model...")
                history = self.train()
                self.plot_history(history)

        train_button.on_click(on_train_button_clicked)

        output = widgets.Output()
        display(ui, output)
```
